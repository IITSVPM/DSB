```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.figure import Figure
```

```python
df1=pd.read_csv('tips_4.csv')
df2=pd.read_csv("AirQuality.csv",encoding='cp1252')
df3=pd.read_csv('dataset_Facebook.csv',sep=";")
```

```python
plt.bar(df1['day'],df1['tip'],color='red')
plt.title('Tips vs day')
plt.xlabel('day')
plt.ylabel('tips')
plt.show()
```

```python
tc=df2.groupby('type')['no2'].sum()
tc
```

```python
tc.plot.pie(label=('type',))
```

```python
sns.lmplot(x='Page total likes',y='like',hue='Paid',data=df3,fit_reg=True,scatter_kws={'alpha':0.6})
plt.legend()

plt.ylim(0,1000)

```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df3['Category'],y=df3['like'],data=df3,palette='viridis')
plt.ylim(0,750)
```

```python
x=df3['Type']
plt.hist(x, bins=4,color='red')
```

```python
sns.histplot(data=df3,x='Type',bins=4,kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Type',data=df3,palette='viridis')
plt.show()
```
