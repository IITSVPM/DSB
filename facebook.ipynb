```python
# shape and reshape by melt and pivot

df_sub6 = df.loc[0:3,['comment', 'like','share','Category']]
df_sub6
```

```python
# melt() is used to go from wide → long format.

# The melt() function reshapes the DataFrame from wide to long format:
# id_vars=['Category'] → keeps the 'Category' column as it is.
# All other columns (comment, like, share) are melted into rows.
# The column named value simply holds the actual data values from the melted columns: 'comment', 'like', and 'share'.

df_sub6.melt(id_vars=['Category'])
```

```python
# index=['Category'] → It groups the data by 'Category' column

# Other numeric columns → Aggregated by mean (by default)

# It's used to summarize large data in a neat table

pd.pivot_table(df_sub6,index=['Category'])
```
