```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Heart.csv')
```

```python
df['AHD'] = df['AHD'].map({'Yes': 1, 'No': 0})
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
print(df.describe())
```

```python
print(df.head(10))
```

```python
X = df[['Age', 'RestBP', 'Chol']]
Y = df['AHD']
```

```python
X.shape
```

```python
X
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=1)
```

```python
Y_train.shape
```

```python
print(X_train.head())
```

```python
Y_test.shape
```

```python
tree_model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=1)
```

```python
X_train.shape
```

```python
print(len(X_train))  # Should print 237
print(len(Y_train))  # Should print 237
```

```python
print(set(Y_train))  # This will show all unique values in Y_train
```

```python
tree_model.fit(X_train, Y_train)
```

```python
dt_pred = tree_model.predict(X_test)
```

```python
print("Actual - Predicted \n")
for x in Y_test:
    for y in dt_pred:
        print(x,"                   ", y)
```

```python
from sklearn import metrics
```

```python
print(metrics.confusion_matrix(Y_test, dt_pred))
```

```python
confusion_matrix = metrics.confusion_matrix(Y_test, dt_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [True, False])

cm_display.plot()
plt.show() 
```

```python
print("Decision Tree Accuracy: %.3f" % accuracy_score(Y_test, dt_pred))
print("Decision Tree F1-Score: %.3f" % f1_score(Y_test, dt_pred))
print("Decision Tree Precision: %.3f" % precision_score(Y_test, dt_pred))
print("Decision Tree Recall: %.3f" % recall_score(Y_test, dt_pred))
```
